# Model Development

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd
```

```python

X = A.drop(['heart_disease_present', 'patient_id'], axis=1)
y = A['heart_disease_present']
```

```python
# StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Data splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# models to evaluate
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(probability=True),
    'Gradient Boosting': GradientBoostingClassifier(),
    'KNN': KNeighborsClassifier()
}
```

```python

results = {}
```

```python

for model_name, model in models.items():
    try:
        print(f"Training {model_name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Check if model supports predict_proba
        if hasattr(model, "predict_proba"):
            y_proba = model.predict_proba(X_test)[:, 1]  # for binary classification
        else:
            y_proba = model.predict(X_test)  
   
        results[model_name] = {
            'Accuracy': accuracy_score(y_test, y_pred),
            'Precision': precision_score(y_test, y_pred),
            'Recall': recall_score(y_test, y_pred),
            'F1-Score': f1_score(y_test, y_pred),
            'AUC-ROC': roc_auc_score(y_test, y_proba)
        }
    except Exception as e:
        print(f"Error in {model_name}: {e}")  # Print any errors encountered for a model
```

```python
#  result in DataFrame 
results_df = pd.DataFrame(results).T
print(results_df)
```

# Why KNN or SVM?

Accuracy: Both KNN and SVM have the highest accuracy (86.11%), meaning they are both great at predicting the correct class in general.

Precision: Both models have perfect precision (1.00), meaning they make no false positives when predicting a positive case.

Recall: Both models have 0.75 recall, meaning they miss a few positive cases, but still do a decent job in identifying the positive class.

AUC-ROC: KNN has a slightly better AUC-ROC (0.87) than SVM (0.84), indicating that KNN performs a little better at distinguishing between the classes across all thresholds.
# Conclusion:
KNN might be slightly better due to its higher AUC-ROC. It strikes a good balance between accuracy, precision, and recall.

However, SVM is also an excellent option, with higher precision and similar recall. It could be better if distinguishing between classes (in terms of separating positives and negatives) is more important.


# Challenges Faced

 Convergence Issues: Models like Logistic Regression occasionally faced convergence issues due to high-dimensional data. Increasing the number of iterations or using different solver helped mitigate these issues.

Feature Engineering: The dataset included many raw numerical features (e.g., cholesterol levels, age), which required careful feature scaling (using StandardScaler) to ensure the models perform optimally.

Model Selection and Hyperparameter Tuning
Choosing the Right Model: Selecting the best model for heart disease prediction can be difficult, as there are many potential algorithms to choose from (e.g., Logistic Regression, SVM, Random Forest, Gradient Boosting).


Model Complexity: More complex models (like Gradient Boosting or SVM) may give better performance but can take longer to train and may be harder to interpret.



Handling Categorical Data
Categorical Features: Many datasets, including heart disease prediction datasets, contain categorical variables (e.g., sex, thalassemia, exercise angina). These need to be properly encoded before being input into the model.

Model Evaluation: Traditional accuracy may not be the best metric when dealing with imbalanced datasets or when the cost of false positives or false negatives is high.
